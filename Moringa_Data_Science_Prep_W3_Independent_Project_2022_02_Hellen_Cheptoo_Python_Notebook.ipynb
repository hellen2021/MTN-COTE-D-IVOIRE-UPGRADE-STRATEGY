## Merge the datasets

```python
#merge the datasets and 
#drop the missing values
data = [df1,df2,df3,df4]
merged_df = pd.concat(data, axis = 1)
df = merged_df.dropna()
df.head()
```

## Data Exploration and Cleaning

```python
df.info()
```

### Drop unnecessary columns

```python
to_drop = ['Unnamed: 0' ,'SIET_ID', 'CELLID', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'LATITUDE', 'LONGITUDE', 'CELL_ID', 'COUNTRY_A', 'COUNTRY_B', 'DW_A_NUMBER', 'DW_B_NUMBER']
df.drop(to_drop, axis = 1, inplace=True)
df.head()
```

### Check for missing values

```python
# Check for null values if merged dataframe
df.isnull().sum()
```

```python
#RENAME COLUMN PRODUTC
df.rename({'PRODUTC': 'PRODUCT'}, axis=1)
```

```python
#REMOVING DUPLICATE COLUMNS TO REMAIN WITH ONE
df.loc[:, ~df.columns.duplicated()]
```

### Split time_date to time and date 

```python
#date
df['date'] = pd.to_datetime(df['DATETIME']).dt.date
```

```python
#time
df['time'] = pd.to_datetime(df['DATETIME']).dt.time
```

**1. Which ones were the most used city for the three days?**

```python
# Getting the count of the column villes('cities')
#
result = pd.value_counts(df['VILLES'])
city_users = result.head()
c_max = result.max()
print(f"The city with the most users appears: {c_max} times\n")
print(f"The Most used cities in 3 days: \n CITY       COUNT\n{city_users}")
```

**2. Which region is most used for the 3 days**

```python
# Getting the region that has most users.
#
region = pd.value_counts(df['REGION'])
print(f"\nThe most used region in 3 days are:\n")
print(f"Region               Count\n\
-------------------------\n{region.head()}")
```

**3. Number of users in Abidjan and those outside of Abidjan.**

```python
# Getting the list of cities in a particular region.
c_loc = pd.value_counts(df['LOCALISATION'])
print(c_loc)
```

**4. Number of Users using Text, Voice and Data in COCODY City.**

```python
# Getting the Services data for MTN users in COCODY city.
# Voice
# data
# sms
#
voice = df[(df['VILLES'] == 'ABOBO') & (df['PRODUCT'] == "Voice")]
data = df[(df['VILLES'] == 'ABOBO') & (df['PRODUCT'] == "data")]
sms = df[(df['VILLES'] == 'ABOBO') & (df['PRODUCT'] == "sms")]

voice_count = pd.value_counts(voice['PRODUCT'])
data_count = pd.value_counts(data['PRODUCT'])
sms_count = pd.value_counts(sms['PRODUCT'])

print("The Number of Users who Accessed MTN Services in \
\n       ABOBO city\n-----------------------------\n\
        In 3 Days\n-----------------------------")
print(f"Total Voice users: {voice_count}\n\nTotal data users: {data_count}\n\nTotal sms users: {sms_count}\n\n")
```

**6. Sum of Billing Price used for the 3 days.**



Billing Price In Cococdy City.

```python
v_cocody = df[df['VILLES'] == 'COCODY']
sum_value = v_cocody['VALUE'].sum()
print(f"The sum of the Billing price per day in Cocody city is\n\
------------------------------------------------------\n{sum_value}")
```

Billing Price In Abobo City.

```python
v_abobo = df[df['VILLES'] == 'ABOBO']
sum_value = v_abobo['VALUE'].sum()
print(f"The sum of the Billing price per day in Abobo city is\n\
------------------------------------------------------\n{sum_value}")
```
